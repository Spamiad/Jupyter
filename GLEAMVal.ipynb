```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
g = xr.open_mfdataset('*.nc')
# g
```

```python
# simple time series
g1d = g.w_frac_B.sel(lat=48.125, lon=11.125)
g1d.plot()
```

```python
g2d = g.w_frac_B.isel(time=0, level1d=0, ens=0)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection=ccrs.Robinson(), aspect='auto')
g2d.plot.imshow(ax=ax, transform=ccrs.PlateCarree(),
                x='lon', y='lat',
                robust=True,
                cmap='PuBu',
                add_colorbar=True,
                extend='both');

plt.show()
```
